```python
# 1) Import the dependencies.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from citipy import citipy
```

```python
# 2) Create a set of random latitude and longitude combinations.

lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)

# 3) Use ZIP to pair both variables into a tuple of arrays.
lat_lngs = zip(lats, lngs)

#### 4) Get the nearest city with Citipy

# 4.1) Create a list form the zipped file.
coords = list(lat_lngs)

# 4.2) Create a list to hold the cities.
cities = []

# 4.3) Use a loop + citypy to get the nearest city.
for coordinate in coords:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)

# 4.4) Nunmber of cities found
len(cities)
```

```python
### 5) API CALL TO OPENWEATHER

# 5.1) Import Dependencies
import requests
from config import weather_api_key
from datetime import datetime

# 5.2) Basic URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# 5.3)  Create an empty list for the data and set a record and set count to 1.
city_data = []
record_count = 1
set_count = 1

# 5.3) Create a for loop to retrieve Data
print("Beginning Data Retrieval     ")
print("-----------------------------")

# 5.4) Loop through all the cities.
for i, city in enumerate(cities):
    
    # 5.4.1) Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50): 
        set_count += 1
        record_count = 1
    
    # 5.4.2) Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")
    
    # 5.4.3) Log the URL, record and set numbers and the city.
    print(f'Processing Record {record_count} of Set {set_count} | {city}')
    
    # 5.4.4) Add 1 to the recorud count.
    record_count += 1
    
    # 5.4.5) Run an API request for each of the cities.
    try:
        # 5.4.6) Parse the JSON and retrieve the data.
        city_weather = requests.get(city_url).json()
        
        # 5.4.7) Parse out the needed data.
        city_country = city_weather["sys"]["country"]
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        weather_desc = city_weather["weather"][0]["description"]
        
        
        # 5.4.8) Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        # 5.4.9) Append the city info
        city_data.append({"City" : city.title(),
                          "Lat" : city_lat,
                          "Lng" : city_lng,
                          "City Weather" : weather_desc,
                          "Max Temp" : city_max_temp,
                          "Humidity" : city_humidity,
                          "Cloudiness" : city_clouds,
                          "Wind Speed" : city_wind,
                          "Country" : city_country,
                          "Date" : city_date})
    
    # 5.4.10) Skip if error.
    except:
        print("City not found. Skipping...")
        pass

# 5.4.11) Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")    
```

```python
# 5.5) Create the DataFrame
city_data_df = pd.DataFrame(city_data)

new_column_order = ["City","Country","Lat","Lng","Max Temp","Humidity","Cloudiness","Wind Speed","City Weather"]
city_data_df = city_data_df[new_column_order]
city_data_df.head(10)
```

```python
# 6) Export to CSV
# 6.1) Create the output file (CSV).
output_data_file = "WeatherPy_Database.csv"

# 6.2) Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
